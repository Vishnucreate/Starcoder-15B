```python
# -*- coding: utf-8 -*-
"""Optimized JavaScript Parsing and Model Integration"""

# Install necessary libraries
!pip install datasets transformers py_mini_racer

import os
from datasets import load_dataset
from py_mini_racer import py_mini_racer
from transformers import AutoTokenizer, AutoModelForCausalLM
import json
import time
import torch

# Step 1: Initialize the JavaScript environment for parsing
print("Initializing JavaScript parser...")
ctx = py_mini_racer.MiniRacer()
ctx.eval("""
    // Load Esprima parser
    var esprima = {
        parse: function(code) {
            return require('esprima').parseScript(code, { loc: true });
        }
    };
""")
print("JavaScript parser initialized.")

# Step 2: Define a function to parse JavaScript using Esprima
def parse_javascript(code_snippet):
    """
    Efficiently parses JavaScript code using the Esprima parser through py_mini_racer.
    Extracts function names and returns a structured representation of the code.
    """
    try:
        # Use Esprima to parse the JavaScript code
        parsed_tree = ctx.call("esprima.parse", code_snippet)

        # Extract function names from the parsed tree
        functions = []
        def extract_functions(node):
            if node.get('type') == 'FunctionDeclaration':
                functions.append(node['id']['name'] if node['id'] else "Anonymous Function")
            for child in node.get('body', {}).get('body', []):
                extract_functions(child)

        extract_functions(parsed_tree)

        return {
            "original_code": code_snippet,
            "parsed_functions": functions,
            "parsed_tree": parsed_tree  # Full Esprima AST
        }
    except Exception as e:
        print(f"Error parsing JavaScript: {e}")
        return {
            "original_code": code_snippet,
            "parsed_functions": [],
            "parsed_tree": "Error parsing code"
        }

# Step 3: Load the dataset and inspect the schema
print("Loading dataset in streaming mode...")
ds = load_dataset("bigcode/the-stack-v2", split="train", streaming=True)

print("Inspecting dataset schema...")
for sample in ds.take(1):  # Fetch one sample for inspection
    print("Sample Schema:", sample)

# Step 4: Filter the dataset for JavaScript samples
start_time = time.time()
print("Filtering dataset for JavaScript...")
javascript_subset = []
try:
    javascript_files = ds.filter(lambda x: x['language'] == 'JavaScript')
    javascript_subset = list(javascript_files.take(100))  # Limit to 100 samples
    print(f"Filtered {len(javascript_subset)} JavaScript samples.")
except Exception as e:
    print(f"Error filtering dataset: {e}")

filter_time = time.time()
print(f"Filtering completed in {filter_time - start_time:.2f} seconds.")

# Step 5: Save the dataset seeds for debugging
print("Saving dataset seeds for debugging...")
try:
    with open("javascript_seeds.json", "w") as f:
        for item in javascript_subset:
            f.write(json.dumps(item) + "\n")
    print("Seeds saved as 'javascript_seeds.json'")
except Exception as e:
    print(f"Error saving seeds: {e}")

# Step 6: Load the model and tokenizer
print("Loading model and tokenizer...")
model_name = "bigcode/tiny_starcoder_py"  # Replace with your model
token = "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"  # Replace with your Hugging Face token

try:
    tokenizer = AutoTokenizer.from_pretrained(model_name, token=token)

    # Set pad_token to eos_token if not already set
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token

    model = AutoModelForCausalLM.from_pretrained(model_name, token=token)
    model.to("cuda" if torch.cuda.is_available() else "cpu")
    print("Model loaded successfully.")
except Exception as e:
    print(f"Error loading model or tokenizer: {e}")

# Step 7: Define a function to generate instruction-response pairs
def generate_instruction_response(seeds, max_new_tokens=256):
    pairs = []
    for i, item in enumerate(seeds):  # Add progress tracking
        print(f"Processing sample {i + 1}/{len(seeds)}...")
        code_snippet = item['content']

        # Parse the JavaScript code
        parsed_data = parse_javascript(code_snippet)

        # Prepare the input for the model
        input_text = (
            f"Analyze the following JavaScript code:\n"
            f"{parsed_data['parsed_tree']}\n\n"
            f"Extracted Functions: {', '.join(parsed_data['parsed_functions'])}"
        )

        inputs = tokenizer(
            input_text,
            return_tensors="pt",
            truncation=True,
            max_length=1024,  # Set max length based on model limits
            padding=True,
            return_attention_mask=True,
        ).to("cuda" if torch.cuda.is_available() else "cpu")

        try:
            # Generate response with max_new_tokens
            outputs = model.generate(
                inputs['input_ids'],
                attention_mask=inputs['attention_mask'],
                max_new_tokens=max_new_tokens,
                pad_token_id=tokenizer.eos_token_id
            )

            # Decode the response
            response = tokenizer.decode(outputs[0], skip_special_tokens=True)

            pairs.append({
                'instruction': input_text,
                'response': response,
                'parsed_data': parsed_data  # Include parsed data for transparency
            })
        except Exception as e:
            print(f"Error generating response for sample {i + 1}: {e}")

    return pairs

# Step 8: Generate instruction-response pairs
response_time = time.time()
print("Generating instruction-response pairs...")
instruction_response_pairs = []
try:
    instruction_response_pairs = generate_instruction_response(javascript_subset)
    print(f"Generated {len(instruction_response_pairs)} instruction-response pairs.")
except Exception as e:
    print(f"Error during instruction-response generation: {e}")

generation_time = time.time()
print(f"Instruction-response generation completed in {generation_time - response_time:.2f} seconds.")

# Step 9: Save results to JSON
output_file = "javascript_instruction_response_pairs_with_parsing.json"
print(f"Saving results to {output_file}...")
try:
    with open(output_file, "w") as f:
        json.dump(instruction_response_pairs, f, indent=4)
    print(f"Results saved to {output_file}.")
except Exception as e:
    print(f"Error saving results: {e}")

save_time = time.time()
print(f"Total runtime: {save_time - start_time:.2f} seconds.")
```
